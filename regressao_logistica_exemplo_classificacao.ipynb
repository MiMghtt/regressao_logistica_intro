```python
#libs necessárias
import numpy as np
import matplotlib.pyplot as plt
```

```python
#função sigmoide
def sigmoid(z):
  return 1 / (1 + np.exp(-z))
```

```python
#dados de exemplo: hoas de estudo e resultado (aprovado ou reprovado)
horas_estudo = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
resultado = np.array([0, 0, 0, 0, 1, 1, 1, 1, 1]) # zero representa reprovado e um representa aprovado
```

```python
plt.scatter(horas_estudo, resultado)
plt.xlabel('Horas de estudo')
plt.ylabel('Resultado (0 = Reprovado, 1 = Aprovado)')
plt.title('Relação entre horas de estudo e resultado')
plt.show()
```

```python
#inicialização dos parâmetros
theta = np.zeros(2) #coeficiente da regressão (inclusive o intercepto)
X = np.column_stack((np.ones(len(horas_estudo)), horas_estudo)) # adicionando coluna para o intercepto
```

```python
#gradiente descendente para encontrar os parâmetros
alpha = 0.01 #taxa de aprendizagem
numero_iteracoes = 1000
m = len(resultado) #tamanho do conjunto de dados
```

```python
for i in range(numero_iteracoes):
  #calculando previsão
  z = np.dot(X, theta)
  h = sigmoid(z)

  #atualizando os parametros
  gradient = np.dot(X.T, (h - resultado)) / m
  theta -= alpha * gradient
```

```python
#plotar linha de decisão
x_values = np.linspace(min(horas_estudo), max(horas_estudo), 100)
y_values = sigmoid(theta[0] + theta[1] * x_values)
plt.scatter(horas_estudo, resultado)
plt.plot(x_values, y_values, color='red')
plt.xlabel('Horas de estudo')
plt.ylabel('Resultado (probabilidade de aprovação)')
plt.title('Regressão logística para previsão de aprovação')
plt.show()
```
